```python
# Удаление ненужных столбцов
data_1 = data_0.drop(['sibsp', 'parch', 'embarked', 'class', 'adult_male', 'deck', 'alive'], axis=1)
data_1.head()
```

```python
data_1.info
```

```python
# Удаление строк с пустыми значениями
data_1.dropna(inplace=True)
data_1.info()
```

```python
#  Проверка на дубликаты
data_1.duplicated().sum()
```

```python
#  Удаление дубликатов
data_1.drop_duplicates(inplace=True)
```

```python
# Минимальный, средний и максимальный возрасты людей на Титанике
data_1.age.min(), data_1.age.mean(), data_1.age.max()
```

```python
# Значение столбца "survived" при условии "age"==0.42
mask = data_1.age==0.42
data_1.loc[mask, "survived"]
```

```python
data_1.loc[mask, :]
```

```python
# Данные по детям до 1 года
mask = data_1.age < 1
data_1.loc[mask, :]
```

```python
# Данные по детям от 10 до 18 лет
mask = (data_1["age"]>=10) & (data_1["age"]<18)
data_1.loc[mask, :]
```
